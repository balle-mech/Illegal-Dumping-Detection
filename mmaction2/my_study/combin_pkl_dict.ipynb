```python
import os
import pickle

pickle_split_dict = {}  # split用
pickle_anno_list = []  # annotation用

for d_type in ["train", "val"]:
    # dtypeごとのpickleファイル名を取得
    pkl_dataset_path = './data/pkl_dataset'
    pickle_dest_root = os.path.join(pkl_dataset_path, str(d_type))
    tmp_list = []
    # print('os.listdir(pkl_dataset_path):', os.listdir(pkl_dataset_path))
    for file_name in os.listdir(pickle_dest_root):
        file_path = os.path.join(pickle_dest_root, file_name)

        if os.path.isfile(file_path):
            # print(file_path)
            # tmp_list.append(file_path) # 絶対パス

            base_n = file_name.rsplit('.', 1)[0]  # 拡張子を除く rspitは右から分割
            tmp_list.append(base_n)  # 相対パス
            # print(base_n)

            # pickleファイル読み込み
            with open(file_path, mode="rb+") as f:
                tmp_dict = pickle.load(f)
                pickle_anno_list.append(tmp_dict)

    # ここでd_typeごとにpickle_split_dictにファイル名だけを格納
    pickle_split_dict['xsub_'+str(d_type)] = tmp_list

print('pickle_split_dict:', pickle_split_dict)
print('pickle_anno_list:', pickle_anno_list)
```

```python
print('pickel_anno_list[1]:', pickle_anno_list[1])
print('type(pickle_anno_list):', type(pickle_anno_list))
print('type(pickle_anno_list[1]):', type(pickle_anno_list[1]))
```

pklファイルを一つにまとめる

```python
pickle_dataset = {
    "split": pickle_split_dict,
    "annotations": pickle_anno_list
}

pickle_save_path = os.path.join(pkl_dataset_path, 'custom_dataset', 'custom_dataset.pkl')

with open(pickle_save_path, 'wb') as p:
    pickle.dump(pickle_dataset, p)
```
